### PreparaciÃ³n de datos

Antes de realizar los cÃ¡lculos, es crucial convertir la columna 'Venta' a un tipo numÃ©rico, ya que actualmente es un `object` y puede contener valores que no son directamente numÃ©ricos. Usaremos `pd.to_numeric` con `errors='coerce'` para convertir cualquier valor problemÃ¡tico a `NaN` (Not a Number), lo que nos permitirÃ¡ operar con la columna.

TambiÃ©n convertiremos las columnas de fecha y hora a formatos adecuados para su anÃ¡lisis.

```python
# Convertir 'Venta' a numÃ©rico. Usar errors='coerce' para manejar valores no numÃ©ricos.
df['Venta'] = pd.to_numeric(df['Venta'], errors='coerce')

# Convertir 'Fecha Vta' a datetime
df['Fecha Vta'] = pd.to_datetime(df['Fecha Vta'], format='%d/%m/%Y', errors='coerce')

# Extraer la hora de 'Hora Vta' si es necesario, o usar directamente para agrupaciÃ³n
df['Hora'] = pd.to_datetime(df['Hora Vta'], format='%H:%M:%S', errors='coerce').dt.hour

print("Tipos de datos actualizados:")
display(df.info())
```

### 1. Venta Total (Global y por Tienda)

Calcularemos la suma total de las ventas para todas las tiendas y luego la suma de ventas para cada tienda individualmente.

```python
import plotly.express as px
import pandas as pd # Import pandas if not already imported in this cell

# Venta total global
venta_total_global = df['Venta'].sum()
print(f"Venta Total Global: {venta_total_global:,.2f}")

# Venta total por tienda
venta_total_por_tienda = df.groupby('Tienda')['Venta'].sum()
print("\nVenta Total por Tienda:")
display(venta_total_por_tienda)

# Generar grÃ¡fico interactivo de barras para Venta Total por Tienda con Plotly Express
fig = px.bar(venta_total_por_tienda.reset_index(),
             x='Tienda',
             y='Venta',
             title='Venta Total por Tienda',
             labels={'Tienda': 'Tienda', 'Venta': 'Venta Total'})
fig.update_xaxes(title_text='Tienda')
fig.update_yaxes(title_text='Venta Total')
fig.show()
```

### 2. NÃºmero de Tickets (Global y por Tienda)

Contaremos el nÃºmero de tickets Ãºnicos emitidos en total y por cada tienda.

```python
import matplotlib.pyplot as plt
import plotly.express as px

# NÃºmero de tickets global
num_tickets_global = df['Ticket'].nunique()
print(f"NÃºmero Total de Tickets: {num_tickets_global}")

# NÃºmero de tickets por tienda
num_tickets_por_tienda = df.groupby('Tienda')['Ticket'].nunique()
print("\nNÃºmero de Tickets por Tienda:")
display(num_tickets_por_tienda)

# Generar grÃ¡fico interactivo de barras para NÃºmero de Tickets por Tienda con Plotly
fig = px.bar(num_tickets_por_tienda,
             x=num_tickets_por_tienda.index,
             y='Ticket',
             title='NÃºmero de Tickets por Tienda',
             labels={'x': 'Tienda', 'Ticket': 'NÃºmero de Tickets'})
fig.update_xaxes(title_text='Tienda')
fig.update_yaxes(title_text='NÃºmero de Tickets')
fig.show()
```

### 3. Venta por DÃ­a (Global y por Tienda)

Analizaremos cÃ³mo se distribuyen las ventas a lo largo de los dÃ­as, tanto en general como para cada tienda.

```python
import plotly.express as px

# Venta por dÃ­a global
venta_por_dia_global = df.groupby('Fecha Vta')['Venta'].sum().sort_index()
print("Venta por DÃ­a (Global):")
display(venta_por_dia_global)

# GrÃ¡fico interactivo de Venta por DÃ­a (Global)
fig_global = px.line(venta_por_dia_global, x=venta_por_dia_global.index, y='Venta', title='Venta por DÃ­a (Global)')
fig_global.update_xaxes(title_text='Fecha')
fig_global.update_yaxes(title_text='Venta Total')
fig_global.show()

# Venta por dÃ­a por tienda
venta_por_dia_por_tienda = df.groupby(['Tienda', 'Fecha Vta'])['Venta'].sum().unstack(level=0)
print("\nVenta por DÃ­a (por Tienda):")
display(venta_por_dia_por_tienda)

# GrÃ¡fico interactivo de Venta por DÃ­a (por Tienda)
# Reshape data for plotly express to handle multiple lines easily
venta_por_dia_por_tienda_melted = venta_por_dia_por_tienda.reset_index().melt(id_vars='Fecha Vta', var_name='Tienda', value_name='Venta')
fig_tienda = px.line(venta_por_dia_por_tienda_melted, x='Fecha Vta', y='Venta', color='Tienda', title='Venta por DÃ­a (por Tienda)')
fig_tienda.update_xaxes(title_text='Fecha')
fig_tienda.update_yaxes(title_text='Venta Total')
fig_tienda.show()
```

### 4. Venta por Hora (Global y por Tienda)

Examinaremos los patrones de venta segÃºn la hora del dÃ­a.

```python
import plotly.express as px

def format_hour(hour):
    if hour == 0:
        return '12 AM'
    elif hour < 12:
        return f'{hour} AM'
    elif hour == 12:
        return '12 PM'
    else:
        return f'{hour - 12} PM'

# Venta por hora global
venta_por_hora_global = df.groupby('Hora')['Venta'].sum().sort_index()
print("Venta por Hora (Global):")
display(venta_por_hora_global)

# GrÃ¡fico interactivo de Venta por Hora (Global)
formatted_hours_global = venta_por_hora_global.index.map(format_hour)
fig_hora_global = px.bar(venta_por_hora_global, x=formatted_hours_global, y='Venta', title='Venta por Hora (Global)')
fig_hora_global.update_xaxes(title_text='Hora del DÃ­a')
fig_hora_global.update_yaxes(title_text='Venta Total')
fig_hora_global.show()

# Venta por hora por tienda
venta_por_hora_por_tienda = df.groupby(['Tienda', 'Hora'])['Venta'].sum().unstack(level=0)
print("\nVenta por Hora (por Tienda):")
display(venta_por_hora_por_tienda)

# GrÃ¡fico interactivo de Venta por Hora (por Tienda)
venta_por_hora_por_tienda_melted = venta_por_hora_por_tienda.reset_index().melt(id_vars='Hora', var_name='Tienda', value_name='Venta')
venta_por_hora_por_tienda_melted['Hora Formato'] = venta_por_hora_por_tienda_melted['Hora'].map(format_hour)
fig_hora_tienda = px.line(venta_por_hora_por_tienda_melted, x='Hora Formato', y='Venta', color='Tienda', title='Venta por Hora (por Tienda)')
fig_hora_tienda.update_xaxes(title_text='Hora del DÃ­a')
fig_hora_tienda.update_yaxes(title_text='Venta Total')
fig_hora_tienda.show()
```

### 5. Venta por Producto (Global y por Tienda)

Obtendremos las ventas totales para cada producto.

```python
import plotly.express as px
import pandas as pd

# Venta por producto global
venta_por_producto_global = df.groupby('Producto')['Venta'].sum().sort_values(ascending=False)
print("Venta por Producto (Global):")
display(venta_por_producto_global)

# Top 10 productos mÃ¡s vendidos globalmente
top_10_global_products = venta_por_producto_global.nlargest(10)

# GrÃ¡fico interactivo para Top 10 Productos MÃ¡s Vendidos (Global)
fig_global_top10_products = px.bar(top_10_global_products.reset_index(), # reset_index to make Producto a column for px.bar
                          x='Producto',
                          y='Venta',
                          title='Top 10 Productos MÃ¡s Vendidos (Global)',
                          labels={'Producto': 'Producto', 'Venta': 'Venta Total'},
                          hover_data={'Producto': True, 'Venta': ':.2f'})
fig_global_top10_products.update_xaxes(title_text='Producto')
fig_global_top10_products.update_yaxes(title_text='Venta Total')
fig_global_top10_products.show()

# Ahora, para el top 10 global, veamos sus ventas por tienda
# Filtrar el DataFrame original para incluir solo los top 10 productos globales
df_top10_global_filtered = df[df['Producto'].isin(top_10_global_products.index)]

# Agrupar por Producto y Tienda para obtener las ventas de cada uno
venta_top10_por_producto_y_tienda = df_top10_global_filtered.groupby(['Producto', 'Tienda'])['Venta'].sum().reset_index()

# Calcular la venta total para cada uno de los top 10 productos globales
total_sales_per_product = df_top10_global_filtered.groupby('Producto')['Venta'].sum().rename('Total Venta Producto').reset_index()

# Unir la venta total del producto al DataFrame de ventas por tienda para calcular el porcentaje
venta_top10_por_producto_y_tienda = pd.merge(venta_top10_por_producto_y_tienda, total_sales_per_product, on='Producto')

# Calcular el porcentaje de contribuciÃ³n de cada tienda a la venta total de cada producto
venta_top10_por_producto_y_tienda['Porcentaje Venta Tienda'] = (venta_top10_por_producto_y_tienda['Venta'] / venta_top10_por_producto_y_tienda['Total Venta Producto']) * 100

# Crear un grÃ¡fico de barras apiladas con hooverdata explÃ­cito
fig_stacked_bar = px.bar(venta_top10_por_producto_y_tienda,
                         x='Producto',
                         y='Venta',
                         color='Tienda',
                         title='Venta de los Top 10 Productos Globales por Tienda',
                         labels={'Producto': 'Producto', 'Venta': 'Venta Total', 'Tienda': 'Tienda'},
                         custom_data=['Producto', 'Tienda', 'Venta', 'Porcentaje Venta Tienda'])

# Definir el hovertemplate para mostrar el porcentaje de forma clara
fig_stacked_bar.update_traces(hovertemplate='<b>Producto</b>: %{customdata[0]}<br>'+
                                            '<b>Tienda</b>: %{customdata[1]}<br>'+
                                            '<b>Venta</b>: %{customdata[2]:,.2f}<br>'+
                                            '<b>% ContribuciÃ³n Tienda</b>: %{customdata[3]:.2f}%<extra></extra>')

fig_stacked_bar.update_layout(xaxis={'categoryorder':'total descending'}) # Ordenar por venta total de producto
fig_stacked_bar.show()
```

### 6. Top 10 Productos MÃ¡s Vendidos (Global y por Tienda)

Identificaremos los 10 productos que han generado mayores ventas.

```python
import plotly.express as px
import pandas as pd

# Top 10 productos mÃ¡s vendidos globalmente
top_10_global = df.groupby('Producto')['Venta'].sum().nlargest(10)
print("Top 10 Productos MÃ¡s Vendidos (Global):")
display(top_10_global)

# GrÃ¡fico interactivo para Top 10 Productos MÃ¡s Vendidos (Global)
fig_global_top10 = px.bar(top_10_global.reset_index(), # reset_index to make Producto a column for px.bar
                          x='Producto',
                          y='Venta',
                          title='Top 10 Productos MÃ¡s Vendidos (Global)',
                          labels={'Producto': 'Producto', 'Venta': 'Venta Total'},
                          hover_data={'Producto': True, 'Venta': ':.2f'})
fig_global_top10.update_xaxes(title_text='Producto')
fig_global_top10.update_yaxes(title_text='Venta Total')
fig_global_top10.show()

# Top 10 productos mÃ¡s vendidos por tienda
top_10_por_tienda = df.groupby(['Tienda', 'Producto'])['Venta'].sum().groupby(level=0, group_keys=False).nlargest(10)
print("\nTop 10 Productos MÃ¡s Vendidos (por Tienda):")
display(top_10_por_tienda)

# GrÃ¡ficos interactivos para Top 10 Productos MÃ¡s Vendidos (por Tienda)
# Iterate through each store to create individual bar charts
for tienda in top_10_por_tienda.index.get_level_values('Tienda').unique():
    productos_tienda = top_10_por_tienda.loc[tienda]

    if not productos_tienda.empty:
        # Convert Series to DataFrame for Plotly Express
        productos_tienda_df = productos_tienda.reset_index()
        fig_tienda_top10 = px.bar(productos_tienda_df,
                                  x='Producto',
                                  y='Venta',
                                  title=f'Top 10 Productos MÃ¡s Vendidos en {tienda}',
                                  labels={'Producto': 'Producto', 'Venta': 'Venta Total'},
                                  hover_data={'Producto': True, 'Venta': ':.2f'})
        fig_tienda_top10.update_xaxes(title_text='Producto')
        fig_tienda_top10.update_yaxes(title_text='Venta Total')
        fig_tienda_top10.show()
    else:
        print(f"No hay datos para Top 10 Productos en {tienda}.")
```
